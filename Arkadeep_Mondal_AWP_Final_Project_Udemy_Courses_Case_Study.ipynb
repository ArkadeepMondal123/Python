Q9) Which subject has the highest average number of subscribers?

```python
a=data.groupby('subject')
b=pd.DataFrame(a['num_subscribers'].mean())
b
```

```python
sns.barplot(x=b.index,y=b['num_subscribers'])
plt.xlabel('Subject')
plt.ylabel('Average number of Subscribers')
plt.title('Average number of subscribers for each subject')
plt.xticks(rotation=90)
plt.show()
```

```python
# We can observe that subject Web Development has the highest average number of subscribers.
```

Q11) Display titles of top 5 courses with highest number of subscribers.

```python
num_max_subscribers=data.groupby('course_title')['num_subscribers'].max()
data1=pd.DataFrame(num_max_subscribers.sort_values(ascending=True),columns=['num_subscribers']).tail(5)[::-1]
data1
```

```python
sns.barplot(x=data1['num_subscribers'],y=data1.index)
plt.title("Courses with highest number of subscribers")
plt.show()
```
